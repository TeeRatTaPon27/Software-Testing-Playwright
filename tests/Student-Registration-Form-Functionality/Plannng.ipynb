# **Planning**

## **Field Definitions**

| ชื่อฟิลด์ (Label)        | ประเภท Element     | รูปแบบข้อมูล (Format)                                   |
|--------------------------|--------------------|----------------------------------------------------------|
| First Name / Last Name   | Text Input         | ตัวอักษร (Alphabetic)                                   |
| Email                    | Text Input         | รูปแบบอีเมล (name@example.com)                          |
| Gender                   | Radio Button       | ชาย, หญิง, อื่นๆ (เลือกได้เพียง 1)                      |
| Mobile                   | Text Input         | ตัวเลข 10 หลัก                                           |
| Date of Birth            | Date Picker        | DD MMM YYYY (เลือกผ่านปฏิทิน)                           |
| Subjects                 | Auto-suggest       | เลือกได้หลายค่า (แสดงเป็น Tags)                         |
| Hobbies                  | Checkbox           | กีฬา, การอ่าน, ดนตรี (เลือกได้หลายข้อ)                   |
| Picture                  | File Upload        | ไฟล์รูปภาพ (jpg, png)                                   |
| Current Address          | Text Area          | ตัวอักษรและตัวเลข (หลายบรรทัด)                         |
| State / City             | Dropdown           | เลือกจากรายการ (City ต้องถูกกรองตาม State)              |

### **Mandatory Fields(ฟิลด์บังคับ)**

- **First Name**
- **Last Name**
- **Gender**
- **Mobile**

### **Fields Relational**

- **State -> City**: 
    - Select **City** from list (Filtered by **State**)
    - **City** จะ **Disabled** ถ้า **State** ยังไม่ถูกเลือก
- **Subjects**: สามารถเลือกใส่ หลายวิชา ได้ และ สามารถลบวิชาเหล่านั้นออกจาก tag ได้
